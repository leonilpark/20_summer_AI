```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow

data = pd.read_csv('롤롤.csv',encoding='euc-kr')

data=data.drop(['gameId','B엘리트몬스터','B전령','B토탈골드','B경험치','BCS','B골드차이','B경험치차이',
            'B분당골드','R와드', 'R와드파괴', 'R퍼블', 'R킬', 'R데스', 'R어시스트',
       'R엘리트몬스터','R전령','R토탈골드','R경험치'],axis=1)


data = data.values

X = data[:,1:]
y = data[:,0]


X_train,X_test,y_train,y_test=\
    train_test_split(X,y,
    test_size=0.4,
    train_size=0.6,
    random_state=42,
    shuffle=True)
```

```python
def dist(x,index_num): 
    return np.sqrt(((index_num-x)**2).sum())

def knn(X,y,index_num,k): 
    vals = []
    m =  X.shape[0]
    
    for i in range(m):
        d = dist(X[i], index_num)
        vals.append([d, y[i]])
       
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1], return_counts=True)
    pred = new_vals[1].argmax()
    return pred

def victory(number):
    if number == 1:
        return "Victory"
    else:
        return "Defeat"
    
    
def predict(X,y,index_num,k): 
    pred = knn(X,y,index_num,k)
    game = victory(int(pred))
    print(f"Predicted Value : {game}")
    return game
    
```

```python
everage = 0
cycle = 100
for index_num in range(cycle):
    pred_game = predict(X_train,y_train,X_test[index_num],k=20)
    game = victory(y_test[index_num])
    print(f"Actual Value : {game}")
    if game == pred_game:
        print("Correct")
        everage +=1
    else:
        print("Incorrect")
    print("---------------")
everage = everage / cycle
print(f"Accuracy : {everage*100}%")
```
